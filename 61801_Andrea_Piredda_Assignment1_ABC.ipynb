By giving another look at this dataset, we can notice how there is some data with always the same lenght of residence (6,801) and the same income (120558.26). These values don't quite follow the pattern of the other ones. Let's investigate more.

```python
dfd[(dfd['LENGTH_OF_RESIDENCE'] == 6.801) & (round(dfd['INCOME'], 2) == 120558.26)]
```

We eventually find out that these values make up almost 10% of the dataset - but every value other than INDIVIDUAL_ID, LENGHT_OF_RESIDENCE AND INCOME always varies. We decide to distribute these values, by filling them with random distributions from the dataset.

```python
import numpy as np
length_of_residence_constant = 6.801
income_constant = 120558.26

# Find the non-constant values to sample from
lor_values_to_sample = dfd.loc[dfd['LENGTH_OF_RESIDENCE'] != length_of_residence_constant, 'LENGTH_OF_RESIDENCE'].dropna()
inc_values_to_sample = dfd.loc[dfd['INCOME'] != income_constant, 'INCOME'].dropna()

# Replace the constant values by randomly sampling from the non-constant values
dfd.loc[dfd['LENGTH_OF_RESIDENCE'] == length_of_residence_constant, 'LENGTH_OF_RESIDENCE'] = np.random.choice(lor_values_to_sample, size=len(dfd[dfd['LENGTH_OF_RESIDENCE'] == length_of_residence_constant]), replace=True)
dfd.loc[dfd['INCOME'] == income_constant, 'INCOME'] = np.random.choice(inc_values_to_sample, size=len(dfd[dfd['INCOME'] == income_constant]), replace=True)
```

## Question 2: Predictive Modeling estimate Costumer Value (35 points)

#### Background: ABC Insurance aims to optimize its marketing and product offerings by predicting which customers are likely to be the most valuable. The CURR_ANN_AMT_CAR, CURR_ANN_AMT_HEALTH, and CURR_ANN_AMT_HOUSE fields in the df_customer dataset can serve as indicators of customer value.

#### Q2.1 Build a baseline model to estimate the costumer value based on their characteristics. (15 points)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import math
```

```python
# summing up the customer value
df['Composite_Customer_Value'] = df['CURR_ANN_AMT_CAR'] + df['CURR_ANN_AMT_HEALTH'] + df['CURR_ANN_AMT_HOUSE']

# drop unnecessary columns to define features DataFrame
features_df = df.drop(['INDIVIDUAL_ID', 'ADDRESS_ID', 'Composite_Customer_Value', 'CURR_ANN_AMT_CAR', 'CURR_ANN_AMT_HEALTH', 'CURR_ANN_AMT_HOUSE', 'CUST_ORIG_DATE', 'DATE_OF_BIRTH', 'SOCIAL_SECURITY_NUMBER', 'CUST_START_DATE', 'MARITAL_STATUS_Single'], axis=1)

cat_features = pd.get_dummies(features_df[['HOME_OWNER','COLLEGE_DEGREE','GOOD_CREDIT',	'ADVANCED_STUDIES']])
scaler = StandardScaler()
num_features = scaler.fit_transform(features_df[['INCOME','HAS_CHILDREN','LENGTH_OF_RESIDENCE','Average_Home_Market_Value']])

features = cat_features + num_features
target = df['Composite_Customer_Value']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

linear_reg_model = LinearRegression()

# Fit the model to the training data
linear_reg_model.fit(X_train, y_train)

# Predict on the test data
y_pred = linear_reg_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)

print(f'Mean Squared Error: {mse}')
print(f'RMSE Score: {rmse}')
```
